```python
import numpy
import sympy
import math

ERROR = 1e-7
LEFT = 2
RIGHT = 3
```

```python
x = sympy.Symbol("x")

func = x * sympy.log(x, 10) - 1

def f(x_axis):
    return sympy.lambdify(x, func)(x_axis)

def opposite_signs(x, y):
    return numpy.signbit(x) != numpy.signbit(y)
```

```python
def bisection(left=LEFT, right=RIGHT, error=ERROR):
    while right - left >= error:
        mid = left + (right - left) / 2.0
        left_value = f(left)
        mid_value = f(mid)
        right_value = f(right)

        if mid_value == 0:
            return mid
        elif opposite_signs(left_value, mid_value):
            right = mid
        elif opposite_signs(mid_value, right_value):
            left = mid

    return mid

print(f"Valor encontrando com o método da bissecção: {bisection()}")
```

```python
def newton_raphson(point):
  f_derivate = func.diff(x)

  new_val = None  
  while True:
    new_val = point - (f(point) / sympy.lambdify(x, f_derivate)(point))
    if abs(new_val - point) <= ERROR:
      break
    point = new_val
    
  return new_val


print(f"Valor encontrando com o método de Newton: {newton_raphson(LEFT)}")
```

```python
def fake_position(x0=LEFT, x1=RIGHT):
    y_x0 = f(x0)
    y_x1 = f(x1)

    next_x = None

    while True:
        next_x = x1 - y_x1 * ((x1 - x0) / (y_x1 - y_x0))
        next_y = f(next_x)
        if abs(next_x - x1) < ERROR:
            break
        x0 = x1
        y_x0 = y_x1

        x1 = next_x
        y_x1 = next_y

    return next_x


print(f"Valor encontrando com o método da fake position: {fake_position()}")
```
