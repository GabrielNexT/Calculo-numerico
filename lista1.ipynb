```python
import numpy
import sympy
import time

ERROR = 1e-7
LEFT = 2
RIGHT = 3
```

```python
x = sympy.Symbol("x")

func = x * sympy.log(x, 10) - 1

def f(x_axis, mathFunc=func):
    return sympy.lambdify(x, mathFunc)(x_axis)

def opposite_signs(x, y):
    return numpy.signbit(x) != numpy.signbit(y)
```

```python
def bisection(left=LEFT, right=RIGHT, error=ERROR):
    iterations = 0
    mid = None
    while right - left >= error:
        iterations += 1
        mid = left + (right - left) / 2.0
        left_value = f(left)
        mid_value = f(mid)
        right_value = f(right)

        if mid_value == 0:
            return mid
        elif opposite_signs(left_value, mid_value):
            right = mid
        elif opposite_signs(mid_value, right_value):
            left = mid

    return mid, f(mid), iterations
```

```python
def newton_raphson(point=LEFT, mathFunc=func):
    f_derivate = mathFunc.diff(x)
    iterations = 0
    new_val = None
    while True:
        iterations += 1
        new_val = point - (
            sympy.lambdify(x, mathFunc)(point) / sympy.lambdify(x, f_derivate)(point)
        )
        if abs(new_val - point) <= ERROR:
            break
        point = new_val

    return new_val, f(new_val), iterations
```

```python
def fake_position(x0=LEFT, x1=RIGHT):
    y_x0 = f(x0)
    y_x1 = f(x1)
    iterations = 0
    next_x = None

    while True:
        iterations += 1
        next_x = x1 - y_x1 * ((x1 - x0) / (y_x1 - y_x0))
        next_y = f(next_x)
        if abs(next_x - x1) < ERROR:
            break
        x0 = x1
        y_x0 = y_x1

        x1 = next_x
        y_x1 = next_y

    return next_x, f(next_x), iterations
```

```python
def secante(x0 = LEFT, x1 = RIGHT):
  iterations = 0
  while abs(x1 - x0) > ERROR:
    iterations += 1
    x2 = x1 - (f(x1) * (x1 - x0)/(f(x1)-f(x0)))
    x0 = x1
    x1 = x2

  return x1, f(x1), iterations
```

### Questão 2

O resultado da questão 2 pode ser encontrado com o retorno do teste abaixo, com ele nós podemos verificar o resultado que cada função encontrou, seu tempo de execução e o número de iterações executados pelo algoritmo.

```python
def printResult(name, method):
  start = time.time()
  x, y, iterations = method()
  end = time.time()

  print(f"Method {name}")
  print(f"(X, Y) = ({x}, {y})")
  print(f"Total time: {end-start:.3f}s")
  print(f"Iterations: {iterations}\n")
  
printResult("Bisection", bisection)
printResult("Newton Raphson", newton_raphson)
printResult("Bisection", fake_position)
printResult("Bisection", secante)
```
